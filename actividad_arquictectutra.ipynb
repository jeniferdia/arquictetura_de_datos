#3)eliminacion y 4)normalizacion

Limpiar valores numéricos (quitar comas y convertir a float)

```python
cols_moneda = ['valor_credito', 'valor_inversion', 'valor_subsidio']
for col in cols_moneda:
    df[col] = df[col].replace({',': ''}, regex=True).astype(float)
```

 Convertir fecha a formato estándar AAAA-MM-DD

```python
df['fecha_credito'] = pd.to_datetime(df['fecha_credito'])
```

Estandarizar textos (quitar espacios extra y pasar a mayúsculas iniciales)

```python
df['sexo'] = df['sexo'].str.strip().str.capitalize()
```

Definir las columnas a eliminar

```python
columnas_a_borrar = ['indicador_lec', 'valor_subsidio', 'nuevo','cod_municipio']
```

ELIMINAR columnas sin información (constantes o en cero)

```python
df = df.drop(columns=columnas_a_borrar)

print("Columnas eliminadas. Ahora el dataset es más limpio.")
```

#Normalización.  transformar una "tabla plana" (donde todo está amontonado y repetido) en un Modelo en Estrella

Creación de "Dimensiones" (Tablas Maestras)

```python
dim_productor = df[['tipo_productor', 'tipo_beneficiario', 'tipo_persona']].drop_duplicates().reset_index(drop=True)
dim_productor['id_productor'] = dim_productor.index + 1
```

El Mapeo (Conectar las tablas)

```python
dim_destino = df[['cod_destino', 'linea_credito', 'cadena', 'eslabon_cadena']].drop_duplicates().reset_index(drop=True)
dim_destino['id_destino_meta'] = dim_destino.index + 1
```

```python
df = df.merge(dim_productor, on=['tipo_productor', 'tipo_beneficiario', 'tipo_persona'])
df = df.merge(dim_destino, on=['cod_destino', 'linea_credito', 'cadena', 'eslabon_cadena'])
```

Creación de la "Tabla de Hechos" (Fact Table)

```python
fact_desembolsos = df[[
    'id_credito', 'fecha_credito', 'id_productor', 'id_destino_meta',
    'sexo', 'valor_credito', 'valor_inversion', 'plazo', 'cuotas'
]]
```

histo

```python
dim_productor.to_csv('dim_productor.csv', index=False)
dim_destino.to_csv('dim_destino.csv', index=False)
fact_desembolsos.to_csv('fact_desembolsos_limpio.csv', index=False)

print("Normalización completada exitosamente.")
```

histograma

```python


# Configurar el estilo para que se vea moderno en la infografía
plt.figure(figsize=(12, 7))
sns.set_style("whitegrid")

# Crear el histograma
# bins=15 crea 15 barras; puedes ajustar este número según veas los datos
ax = sns.histplot(df['valor_credito'], bins=15, color='#2E7D32', edgecolor='white', alpha=0.8)

# Personalización de títulos
plt.title('¿De cuánto son los créditos en Guática?', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Monto del Crédito (Pesos)', fontsize=12)
plt.ylabel('Número de Personas/Créditos', fontsize=12)

# Formatear el eje X para que los números tengan puntos de mil (5.000.000 en lugar de 5000000)
import matplotlib.ticker as ticker
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

# Eliminar bordes innecesarios para un look más limpio
sns.despine()

plt.show()
```

```python


# Configurar el estilo visual
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Crear el histograma del valor del crédito
# bins=20 divide los montos en 20 grupos para ver mejor la forma
ax = sns.histplot(df['valor_credito'], bins=20, kde=True, color='forestgreen')

# Personalización para la infografía
plt.title('Distribución de Montos de Crédito en Guática', fontsize=15)
plt.xlabel('Valor del Crédito (Pesos)', fontsize=12)
plt.ylabel('Cantidad de Créditos', fontsize=12)

# Ajustar formato de moneda en el eje X
import matplotlib.ticker as ticker
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

plt.show()
```
